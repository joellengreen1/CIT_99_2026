# Week 2 Lecture 
During this week you will learn to perform **for loops** and **create functions**.  Let's start with the for loop.  Sometimes you want to repeat a statement or block of statements a number of times.  When this occurs, we create a for loop to complete the process.  
```
for index in range(start, end+1):
    block of code
```

Notice the range(start, end+1) function. It assigns a value for the index from start to end, incrementing by one from the start to the end.  On the first statement notice it ends in : (colon).  The range(n) function will start at zero and end at n-1.
## Example of for-loop printing out value 0 to 20
```
for i in range(0, 21):
    print(i)
```
Like the if-statement, we indent the block of code to repeat.  Also, we include an index that will take on values 0 to end.  In our case, the index is i.  The code above with change the i values from 0 to 20 and after changing the i value it will print out i.
                                           

`````python
for i in range(0, 21):
    print(i)
`````

# local and global variables
local variables are defined within a function but cannot be accessed outside the function.  A variable that has been defined in the program but not in the function, that variable is global.

## Scope Identification
Use these four rules to tell whether a variable belongs to a local scope or the global scope:

1.A variable in the global scope (that is, outside all functions) is always a global variable.

2.A variable in a function with a global statement is always a global variable in that function.

3.Otherwise, if a function uses a variable in an assignment statement, it is a local variable.

4.However, if the function uses a variable but never in an assignment statement, it is a global variable.

`````python
def spam():
    global eggs
    eggs = 'spam'  # This is the global variable.
    print('eggs defined in spam() ',eggs)

def bacon():
    eggs = 'bacon'  # This is a local variable.
    print('eggs defined in bacon() ',eggs)

def ham():
    print('eggs defined in ham() ',eggs)  # This is the global variable.
    
eggs = 'global'  # This is the global variable.
print('global >',eggs)
spam()
bacon()
ham()
print('global >',eggs)
`````
